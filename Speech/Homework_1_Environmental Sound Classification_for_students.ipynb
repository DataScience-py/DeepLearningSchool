```python
from random import sample
from matplotlib.colors import same_color
import torch
import torchaudio
from torch.utils.data import Dataset
import pandas as pd
import os
import random
from pathlib import Path

class SimpleAudioDataset(Dataset):
    """A dataset to load, preprocess, and augment audio files."""

    def __init__(self, data, audio_dir: str, do_augmentation: bool = False, sample_rate: int = 16000, target_len: int = 80000, noise_factor: float = 0.007) -> None:
        self.data = data
        self.audio_dir = audio_dir
        self.do_augmentation = do_augmentation
        self.sample_rate = sample_rate
        self.target_len = target_len
        self.noise_factor = noise_factor
        self.max_shift_samples = int(sample_rate * 0.1)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        data = self.data.iloc[index]
        # 1. Get the file path and label
        audio_file_path = Path(self.audio_dir) / data[0]
        label = data[1]

        # 2. LOAD: Load the raw audio file
        # here is your code
        signal, sample_rate = torchaudio.load(audio_file_path)

        # 3. PREPROCESS: Apply the preprocessing steps
        signal = self._resample(signal, sample_rate) # Resample to the sample rate 16000
        signal = self._stereo_to_mono(signal) # Converts (channels, samples) -> (samples,)
        signal = self._cut_or_pad(signal) # State fixed length

        # 4. AUGMENT: Apply augmentations only if training
        if self.do_augmentation:
            signal = self._augmentation(signal)

        signal = signal.squeeze(1)

        # 5. RETURN: We now have a clean, standardized waveform and its label
        return signal, label

    # --- The Core Preprocessing Functions ---
    def _resample(self, signal, original_sr):
        signal = torchaudio.functional.resample(signal, orig_freq=original_sr, new_freq=self.sample_rate)
        return signal

    def _stereo_to_mono(self, signal):
        # here is your code
        signal = torch.mean(signal, dim=0, keepdim=True)
        return signal

    def _cut_or_pad(self, signal):
        num_channels, sig_len = signal.shape

        if sig_len > self.target_len:
            signal = signal[:, :self.target_len]

        elif sig_len < self.target_len:
            padding_needed = self.target_len - sig_len

            padding = torch.zeros(num_channels, padding_needed, dtype=signal.dtype, device=signal.device)


            signal = torch.cat([signal, padding], dim=1)

        return signal

    def _augmentation(self, signal):
        # here is your code
        if random.random() < 0.1:
            shift = random.randint(-self.max_shift_samples, self.max_shift_samples)
            signal = torch.roll(signal, shifts=(shift,), dims=(1,))

        if random.random() < 0.1:
            noise = torch.randn_like(signal) * self.noise_factor
            signal = signal + noise


        if random.random() < 0.1:
            gain = random.uniform(0.8, 1.2)
            signal = signal * gain # эксперемент 1


        return signal
```

### **Part 2: Building a Model that Learns from Waveforms**

В этом разделе вам нужно написать архитектуру по вашему выбору, которая будет решать задачу классификации на 5 классов.

```python
import torch.nn as nn
import torch.nn.functional as F

class SoundClassificatonModel(nn.Module):
    def __init__(self, input_size=16000, num_classes=5):
        super().__init__()

        # --- Сверточные слои (Feature Extractor) ---
        self.conv1 = nn.Conv1d(1, 16, kernel_size=64, stride=4, padding=32)
        self.pool1 = nn.MaxPool1d(kernel_size=4)
        self.conv2 = nn.Conv1d(16, 32, kernel_size=32, stride=2, padding=16)
        self.pool2 = nn.MaxPool1d(kernel_size=4)
        self.conv3 = nn.Conv1d(32, 64, kernel_size=16, stride=1, padding=8)
        self.pool3 = nn.MaxPool1d(kernel_size=4)
        self.conv4 = nn.Conv1d(64, 128, kernel_size=16, stride=1, padding=8)
        self.pool4 = nn.MaxPool1d(kernel_size=4)
        self.dropout = nn.Dropout(p=0.5)
        self.dropout_cnn = nn.Dropout(p=0.3)
        with torch.no_grad():
            dummy_input = torch.randn(1, 1, input_size)
            final_feature_size = self._get_conv_output_size(dummy_input)

        # --- Полносвязные слои (Classifier) ---
        self.fc1 = nn.Linear(final_feature_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, num_classes)

    def _get_conv_output_size(self, x):
        """Прогон через сверточные слои для определения размера."""
        x = self.pool1(F.relu(self.conv1(x)))
        x = self.pool2(F.relu(self.conv2(x)))
        x = self.pool3(F.relu(self.conv3(x)))
        x = self.pool4(F.relu(self.conv4(x)))
        return x.numel()

    def forward(self, x):
        # x должен быть (N, L), где N - размер батча, L - длина.
        # Добавляем размерность канала: (N, 1, L)
        if x.dim() == 2:
            x = x.unsqueeze(1)
        elif x.dim() != 3:
             raise ValueError("Входной тензор должен быть 2D (N, L) или 3D (N, C, L)")

        # Сверточные блоки
        x = self.pool1(F.relu(self.conv1(x)))
        x = self.pool2(F.relu(self.conv2(x)))
        x = self.pool3(F.relu(self.conv3(x)))
        x = self.pool4(F.relu(self.conv4(x)))

        # Выпрямление
        x = torch.flatten(x, 1)

        # Полносвязные слои
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)

        return x
```

### **Part 3: Training and Evaluation**

В этом разделе вам нужно написать код тренировки и запустить саму тренировку и вывести лучшие значения метрики качества на train и valid данных. Для вашего удобства написана функция отображения значений лоссов и метрики accuracy.

```python
def plot_metrics(train_losses, train_accuracies, test_losses, test_accuracies):
    """
    Plot training and validation metrics
    """
    epochs = range(1, len(train_losses) + 1)

    # Create subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot loss
    ax1.plot(epochs, train_losses, 'b-', label='Training Loss', linewidth=2)
    ax1.plot(epochs, test_losses, 'r-', label='Test Loss', linewidth=2)
    ax1.set_title('Training and Test Loss')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot accuracy
    ax2.plot(epochs, train_accuracies, 'b-', label='Training Accuracy', linewidth=2)
    ax2.plot(epochs, test_accuracies, 'r-', label='Valid Accuracy', linewidth=2)
    ax2.set_title('Training and Valid Accuracy')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy (%)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # Adjust layout and display
    plt.tight_layout()
    plt.show()
```

```python
from asyncio import base_tasks
from cProfile import label
from tabnanny import verbose
from networkx import voronoi_cells
from torch.utils.data import DataLoader
from IPython.display import clear_output
import torch
import torch.nn as nn
from tqdm.auto import tqdm

train_data = train_dataset

valid_data = valid_dataset

UNIQUE_LABELS = ['dog', 'rain', 'crying_baby', 'clock_tick', 'sneezing', 'door_knock', "s"] # класс s - для тестового датасета (в любом случае не участвует)

label_to_index = {label: index for index, label in enumerate(UNIQUE_LABELS)}

index_to_label = {index: label for index, label in enumerate(UNIQUE_LABELS)}

print(f"Словарь меток: {label_to_index}")

def custom_collate_fn(batch):
    """
    Функция сопоставления для DataLoader, которая обрабатывает нестандартный формат:
    ([тензор_сигналов], кортеж_строк_меток)

    :param batch: Список из (data_item, label_item) пар, где data_item - это [tensor].
    :return: Кортеж из двух тензоров: (signals_batch, labels_batch)
    """

    signal_tensors = [data[0][0] for data in batch]

    signals_batch = torch.stack(signal_tensors)



    label_strings = [data[1] for data in batch]

    label_indices = [label_to_index[label] for label in label_strings]

    labels_batch = torch.tensor(label_indices, dtype=torch.long)


    return signals_batch, labels_batch

train_loader = DataLoader(train_data, batch_size=32, collate_fn=custom_collate_fn, shuffle=True, pin_memory=True)

valid_loader = DataLoader(valid_data, batch_size=32, collate_fn=custom_collate_fn, shuffle=False)

device = "cuda" if torch.cuda.is_available() else "cpu"

model = SoundClassificatonModel(input_size=80000).to(device)

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=2e-4)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)


train_losses = []
train_accuracies = []
valid_losses = []
valid_accuracies = []

n_epochs = 30
for epoch in range(n_epochs):
    epoch_train_losses = []
    epoch_train_accuracies = []
    epoch_valid_losses = []
    epoch_valid_accuracies = []

    # Train
    model.train()

    train_loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{n_epochs} (Train)")

    for signals, labels in train_loop:

        signals, labels = signals.to(device), labels.to(device)

        optimizer.zero_grad()

        predictions = model(signals)
        train_loss = criterion(predictions, labels)


        _, predicted_classes = torch.max(predictions.data, 1)
        correct_predictions = (predicted_classes == labels).sum().item()
        train_accuracy_batch = correct_predictions / labels.size(0)

        epoch_train_losses.append(train_loss.item())
        epoch_train_accuracies.append(train_accuracy_batch)

        train_loss.backward()
        optimizer.step()

        train_loop.set_postfix(loss=train_loss.item(), acc=f"{train_accuracy_batch:.4f}")

    # Evaluation
    model.eval()

    valid_loop = tqdm(valid_loader, desc=f"Epoch {epoch+1}/{n_epochs} (Valid)")

    with torch.no_grad():
        for signals, labels in valid_loop:
            signals, labels = signals.to(device), labels.to(device)

            predictions = model(signals)
            valid_loss = criterion(predictions, labels)


            _, predicted_classes = torch.max(predictions.data, 1)
            correct_predictions = (predicted_classes == labels).sum().item()
            valid_accuracy_batch = correct_predictions / labels.size(0)

            epoch_valid_losses.append(valid_loss.item())
            epoch_valid_accuracies.append(valid_accuracy_batch)

            valid_loop.set_postfix(loss=valid_loss.item(), acc=f"{valid_accuracy_batch:.4f}")


    epoch_train_loss = np.mean(epoch_train_losses)
    epoch_train_acc = np.mean(epoch_train_accuracies)
    epoch_valid_loss = np.mean(epoch_valid_losses)
    epoch_valid_acc = np.mean(epoch_valid_accuracies)

    train_losses.append(epoch_train_loss)
    train_accuracies.append(epoch_train_acc)
    valid_losses.append(epoch_valid_loss)
    valid_accuracies.append(epoch_valid_acc)

    scheduler.step(epoch_valid_loss)
    print(f"\n--- Эпоха {epoch+1} ---")
    print(f"Потери (Train/Valid): {epoch_train_loss:.4f} / {epoch_valid_loss:.4f}")
    print(f"Точность (Train/Valid): {epoch_train_acc:.4f} / {epoch_valid_acc:.4f}")

    plot_metrics(train_losses, train_accuracies, valid_losses, valid_accuracies)
    clear_output(wait=True)
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from IPython.display import Audio, display

class ESC50TestDemo:
    def __init__(self, model, test_dataset, device, class_names):
        self.model = model
        self.test_dataset = test_dataset
        self.device = device
        self.classes = class_names
        self.model.eval()

    def predict_audio(self, signal):
        """Predict class for a single audio signal"""
        with torch.no_grad():
            signal = signal.unsqueeze(0).to(self.device)  # Add batch dimension
            outputs = self.model(signal)
            probabilities = torch.softmax(outputs, dim=1)
            confidence, predicted = torch.max(probabilities, 1)

        return predicted.item(), confidence.item(), probabilities.cpu().numpy()[0]

    def run_interactive_demo(self, num_examples=1):
        """Run interactive demo with random test examples"""
        print("ESC-50 Audio Classification Demo!")
        print("=" * 60)

        indices = np.random.choice(len(self.test_dataset), num_examples, replace=False)

        for i, idx in enumerate(indices):
            data_item = self.test_dataset[idx]

            signal = data_item[0][0]
            true_label_string = data_item[1]

            true_label_idx = label_to_index[true_label_string]

            true_class = self.classes[true_label_idx]

            predicted_idx, confidence, all_probs = self.predict_audio(signal)
            predicted_class = self.classes[predicted_idx]


            fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

            ax1.plot(signal.squeeze().numpy())
            ax1.set_title(f'Audio Waveform - Example {i+1}/{num_examples}')
            ax1.set_xlabel('Samples')
            ax1.set_ylabel('Amplitude')
            ax1.grid(True, alpha=0.3)

            colors = ['lightcoral', 'lightgreen']
            correct = predicted_class == true_class
            ax2.barh([0, 1], [confidence * 100, (1-confidence) * 100],
                             color=colors[correct], alpha=0.7)
            ax2.set_yticks([0, 1])
            ax2.set_yticklabels([f'Predicted: {predicted_class}',
                                 f'True: {true_class}'])
            ax2.set_xlabel('Confidence (%)')
            ax2.set_title(f'Prediction ({"✓ Correct" if correct else "✗ Wrong"})')

            top5_indices = np.argsort(all_probs)[-5:][::-1]
            top5_classes = [self.classes[idx] for idx in top5_indices]
            top5_probs = all_probs[top5_indices]

            colors = ['lightgreen' if cls == true_class else 'lightcoral' for cls in top5_classes]
            ax3.barh(range(5), top5_probs * 100, color=colors, alpha=0.7)
            ax3.set_yticks(range(5))
            ax3.set_yticklabels(top5_classes)
            ax3.set_xlabel('Probability (%)')
            ax3.set_title('Top-5 Predictions')
            ax3.invert_yaxis()

            plt.tight_layout()
            plt.show()

            print(f"Playing: {true_class}")
            display(Audio(signal.squeeze().numpy(), rate=16000))

            print(f"Prediction: {predicted_class} ({confidence:.2%})")
            print(f"True label: {true_class}")
            print(f"Correct: {correct}")
            print("=" * 60)

    def evaluate_test_set(self):
        """Evaluate on entire test set"""
        test_loader = DataLoader(self.test_dataset, batch_size=32, shuffle=False, collate_fn=custom_collate_fn)
        self.model.eval()
        correct = 0
        total = 0
        all_predictions = []
        all_labels = []

        with torch.no_grad():
            for data, target in test_loader:
                data, target = data.to(self.device), target.to(self.device)
                outputs = self.model(data)
                _, predicted = torch.max(outputs.data, 1)

                total += target.size(0)
                correct += (predicted == target).sum().item()

                all_predictions.extend(predicted.cpu().numpy())
                all_labels.extend(target.cpu().numpy())

        accuracy = 100 * correct / total
        print(f"ESC-50 Test Set Evaluation (Fold 5):")
        print(f"Correct: {correct}/{total}")
        print(f"Accuracy: {accuracy:.2f}%")

        return accuracy, all_predictions, all_labels

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
demo = ESC50TestDemo(model, valid_dataset, device, UNIQUE_LABELS)
demo.run_interactive_demo(num_examples=5)
test_accuracy, predictions, true_labels = demo.evaluate_test_set()
```
